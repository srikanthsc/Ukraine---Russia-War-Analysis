This dataset describes **Equipment Loss**, **Death Toll**, **Military Wounded** & **Prisoner** of Russia in 2022 Ukraine-Russia War.
All data are official and additionally structured. This dataset was created after 7 days of starting the War.

**Acronyms Used :**
* POW - Prisoner Of War
* MRL - Multiple Rocket Launcher
* APC - Armored Personnel Carrier
* SRBM - Short Range Ballistic Missile

<a id="6"></a>
# <p style="padding:10px;background-color:#bc4749;margin:0;color:#e9c46a;font-family:newtimeroman;font-size:100%;text-align:center;border-radius: 15px 50px;overflow:hidden;font-weight:500">Directions Of Greatest Loss</p>

Let's extract all the directions separately and correct wrongly labelled entries.

```python
great_loss = equipment_dataset[["greatest losses direction"]]
great_loss["greatest losses direction"].replace(np.nan,"NA",inplace=True)
great_loss = great_loss[great_loss["greatest losses direction"].isin(["NA"])==False]
great_loss["greatest losses direction"] = great_loss["greatest losses direction"].str.split(", | , | ,| and ")
great_loss = great_loss.explode("greatest losses direction")
great_loss.replace(to_replace = ["Zaporizhzhya","Slov'yansʹk"], value = ["Zaporizhzhia","Sloviansk"],inplace=True)


print(f'\nGreat losses happened {great_loss.shape[0]} times in {len(great_loss["greatest losses direction"].unique())} different directions.')
print(f'Great losses directions are :\n\n{great_loss["greatest losses direction"].unique()}')

print(f'\nOccurances of direction-wise losses :')
great_loss["greatest losses direction"].value_counts().rename_axis("Greatest Losses Directions").to_frame("Total").T.style.set_properties(**{"background-color": "#bc4749","color":"#e9c46a","border": "1.5px solid black"})
```

```python
plt.subplots(figsize=(20, 8))
p=sns.countplot(x = great_loss["greatest losses direction"], order = great_loss["greatest losses direction"].value_counts(ascending=False).index[:10],palette=pal, saturation=1, edgecolor = "#1c1c1c", linewidth = 3)
p.axes.set_title("\nDirectionwise Occurances\n",fontsize=25)
plt.ylabel("Total Occurances",fontsize=20)
plt.xlabel("\nDirections Of Greatest Loss",fontsize=20)
plt.xticks(rotation=90)
for container in p.containers:
    p.bar_label(container,label_type = "center",padding = 10,size = 25,color = "black",rotation = 0,
    bbox={"boxstyle": "round", "pad": 0.2, "facecolor": "orange", "edgecolor": "white", "linewidth" : 2, "alpha": 1})

    
sns.despine(left=True, bottom=True)
plt.show()
```

**Insights:**

* **Donetsk** is being attacked heavily followed by **Bakhmut**, **Kramatorsk**, **Kryvyi Rih** and other directions.
* Frequencies of attacking in **Popasna** and **Slobozhanskyi** directions are minimum. 

<a id="7"></a>
# <p style="padding:10px;background-color:#bc4749;margin:0;color:#e9c46a;font-family:newtimeroman;font-size:100%;text-align:center;border-radius: 15px 50px;overflow:hidden;font-weight:500">Final Dataset For Deeper Exploration</p>

Let's drop the **"date"** and **"greatest losses direction"** columns from the final dataset as we don't need these colummns anymore for upcoming explorations. And merge **"military auto"**, **"fuel tank"** and **"vehicles and fuel tanks"** columns as these two columns were merged into **"vehicles and fuel tanks"** since 2022-05-01.

```python
war_data = equipment_dataset[:]
war_data["vehicles and fuel tanks"] = war_data[["military auto","fuel tank","vehicles and fuel tanks"]].sum(axis=1)
war_data.drop(columns=["date","greatest losses direction","military auto","fuel tank"],inplace=True)
print(f"After processing, let's have a glimpse of the dataset.")
war_data.head(10).style.set_properties(**{"background-color": "#bc4749","color":"#e9c46a","border": "1.5px solid black"})
```

```python
print("Null Values Of The Dataset :")
war_data.isna().sum().to_frame().T.style.set_properties(**{"background-color": "#bc4749","color":"#e9c46a","border": "1.5px solid black"})
```

So we can see there are some null values for **"special equipment"**, **"mobile SRBM system"** and **"cruise missiles"** columns of the dataset. Let's replace the null values with **"0"** and check if there exists any null values yet and replacing null values of **"POW"** by **"496"** as it is in cumulative sum format.

```python
war_data["POW"].fillna(method ="ffill", inplace = True)
war_data.replace(np.nan,0,inplace=True)
print("After Processing, Null Values Of The Dataset :")
war_data.isna().sum().to_frame().T.style.set_properties(**{"background-color": "#bc4749","color":"#e9c46a","border": "1.5px solid black"})
```

<a id="8"></a>
# <p style="padding:10px;background-color:#bc4749;margin:0;color:#e9c46a;font-family:newtimeroman;font-size:100%;text-align:center;border-radius: 15px 50px;overflow:hidden;font-weight:500">Invaders Destiny</p>

We can see all the values of the dataset are recorded in column-wise cumulative sum format. Let's create a new dataset named **"person_data"** and reformat the datas in daywise record format. 

```python
person_data = war_data[["day","personnel","POW"]]
person_data[["personnel","POW"]] = person_data[["personnel","POW"]].diff().fillna(person_data[["personnel","POW"]])
person_data.set_index("day", drop=True, inplace=True)

print("After processing, let's have a glimpse of the dataset :")
person_data.head(20).astype(int).T.style.set_properties(**{"background-color": "#bc4749","color":"#e9c46a","border": "1.5px solid black"})
```

```python
print(f'Total amount of dead invaders : {person_data["personnel"].sum().astype(int)}')
print(f'Total amount of prisoners : {person_data["POW"].sum().astype(int)}')
```

```python
plt.subplots(figsize=(12, 12))

labels = "Dead Personnel","Prisoner"
size = 0.5

wedges, texts, autotexts = plt.pie([person_data["personnel"].sum(),
                                    person_data["POW"].sum()],
                                    explode = (0,0),
                                    textprops=dict(size= 20, color= "white"),
                                    autopct="%.2f%%", 
                                    pctdistance = 0.75,
                                    radius=1, 
                                    colors = ["#bc4749","#FFC300"],
                                    # labels = labels, # labeldistance = 1.05,
                                    shadow = True,
                                    wedgeprops=dict(width = size, edgecolor = "#1c1c1c", linewidth = 4),
                                    startangle = 1)

plt.legend(wedges, labels, title="Category",loc="center left",bbox_to_anchor=(1, 0, 0.5, 1), edgecolor = "#1c1c1c")
plt.title("\nDestiny Of Invaders",fontsize=20)
plt.show()
```

```python
plt.subplots(figsize=(20, 8))
p = sns.histplot(person_data["personnel"],color="#bc4749",kde=True,bins=50,alpha=1,fill=True,edgecolor=palette_ex[14])
p.axes.lines[0].set_color(palette[6])
p.axes.set_title("\nPersonnel Death Quantity Distribution\n",fontsize=25)
plt.ylabel("Count",fontsize=20)
plt.xlabel("\nPersonnel Amount",fontsize=20)
sns.despine(left=True, bottom=True)

plt.show()
```

```python
print("Frequency of prisoners in a day :")
person_data[person_data["POW"]>0]["POW"].astype(int).value_counts().rename_axis("Prisoner").to_frame("Occurance").T.style.set_properties(**{"background-color": "#bc4749","color":"#e9c46a","border": "1.5px solid black"})
```

```python
plt.subplots(figsize=(20, 8))
p = sns.histplot(person_data[person_data["POW"]>0]["POW"],color="#fca311",kde=True,bins=30,alpha=1,fill=True,edgecolor=palette_ex[14])
p.axes.lines[0].set_color("#bc4749")
p.axes.set_title("\nPrisoner Of War Quantity Distribution\n",fontsize=25)
plt.ylabel("Count",fontsize=20)
plt.xlabel("\nPrisoner Amount",fontsize=20)

sns.despine(left=True, bottom=True)
plt.show()
```

**Insights:**

* Total amount of dead invaders are 59080 which is the fate of 99.17% of captured invaders
* Total amount of prisoners are 496 which is the fate of 0.83% of captured invaders
* Less than 500 invaders were being killed in a day more frequently
* Less than 25 invaders were being sent to prison in a day more frequently

Let's explore overall air equipments destructions. For this we will count **aircraft**, **helicopter**, **drone** and **mobile SRBM system** as air equipments.

```python
air_data = war_data[["day","aircraft","helicopter","drone","mobile SRBM system"]]
air_data[["aircraft","helicopter","drone","mobile SRBM system"]] = air_data[["aircraft","helicopter","drone","mobile SRBM system"]].diff().fillna(air_data[["aircraft","helicopter","drone","mobile SRBM system"]])
air_data.set_index("day" , drop=True, inplace =True)

print("Day-wise destructions of air equipments :")
air_data.head(25).astype(int).T.style.set_properties(**{"background-color": "#bc4749","color":"#e9c46a","border": "1.5px solid black"})
```

```python
_, axs = plt.subplots(2,2,figsize=(20,10),sharex=True)
plt.tight_layout(pad=4.0)

sns.lineplot(x="day",y="aircraft",data=air_data,ax=axs[0,0],color=palette[5])
axs[0,0].set_title("Aircraft\n",fontsize=25)
axs[0,0].set_xlabel("Days",fontsize=20)
axs[0,0].set_ylabel("Quantity",fontsize=20)

sns.lineplot(x="day",y="helicopter",data=air_data,ax=axs[0,1],color=palette[6])
axs[0,1].set_title("Helicopter\n",fontsize=25)
axs[0,1].set_xlabel("Days",fontsize=20)
axs[0,1].set_ylabel("Quantity",fontsize=20)

sns.lineplot(x="day",y="drone",data=air_data,ax=axs[1,0],color=palette[0])
axs[1,0].set_title("Drone\n",fontsize=25)
axs[1,0].set_xlabel("\nDays",fontsize=20)
axs[1,0].set_ylabel("Quantity",fontsize=20)

sns.lineplot(x="day",y="mobile SRBM system",data=air_data[air_data["mobile SRBM system"]>= 0],ax=axs[1,1],color=palette[1])
axs[1,1].set_title("Mobile SRBM System\n",fontsize=25)
axs[1,1].set_xlabel("\nDays",fontsize=20)
axs[1,1].set_ylabel("Quantity",fontsize=20)

plt.suptitle("Daywise Air Equipments Destruction",fontsize=30, y=1.05)

sns.despine(left=True, bottom=True)
plt.show()
```

```python
plt.subplots(figsize=(12, 12))

labels = "Aircraft","Helicopter","Drone"
size = 0.5

wedges, texts, autotexts = plt.pie([air_data["aircraft"].sum(),
                                    air_data["helicopter"].sum(),
                                    air_data["drone"].sum()],
                                    explode = (0,0,0),
                                    textprops=dict(size= 20, color= "white"),
                                    autopct="%.2f%%", 
                                    pctdistance = 0.70,
                                    radius=.9, 
                                    colors = ["#a3a380","#fee440","#bc4749"], 
                                    # labels = labels, # labeldistance = 1.05,
                                    shadow = True,
                                    wedgeprops=dict(width = size, edgecolor = "#1c1c1c", linewidth = 4),
                                    startangle = -67)

plt.legend(wedges, labels, title="Category",loc="center left",bbox_to_anchor=(1, 0, 0.5, 1), edgecolor = "#1c1c1c")
plt.title("\nAir Equipments Destruction",fontsize=20)
plt.show()
```

# <p style="padding:10px;background-color:#bc4749;margin:0;color:#e9c46a;font-family:newtimeroman;font-size:100%;text-align:center;border-radius: 15px 50px;overflow:hidden;font-weight:500">Field Equipments Destiny</p>

Let's explore overall field equipments destructions. For this we will count **tank**, **APC**, **MRL**, **field artillery**, **anti-aircraft warfare** and **vehicles and fuel tanks** as field equipments.
As **vehicles and fuel tanks** has negative values, let's set them to zero.

```python
field_data = war_data[["day","tank","APC","MRL","field artillery","anti-aircraft warfare","vehicles and fuel tanks"]]
field_data[["tank","APC","MRL","field artillery","anti-aircraft warfare","vehicles and fuel tanks"]] = field_data[["tank","APC","MRL","field artillery","anti-aircraft warfare","vehicles and fuel tanks"]].diff().fillna(field_data[["tank","APC","MRL","field artillery","anti-aircraft warfare","vehicles and fuel tanks"]])
field_data[field_data["vehicles and fuel tanks"] < 0]["vehicles and fuel tanks"] = 0
field_data.set_index("day" , drop=True, inplace =True)

print("Day-wise destructions of field equipments :")
field_data.head(25).astype(int).T.style.set_properties(**{"background-color": "#bc4749","color":"#e9c46a","border": "1.5px solid black"})
```

```python
_, axs = plt.subplots(2,3,figsize=(25,12),sharex=True)
plt.tight_layout(pad=4.0)

sns.lineplot(x="day",y="tank",data=field_data,ax=axs[0,0],color=palette[5])
axs[0,0].set_title("\nTank\n",fontsize=25)
axs[0,0].set_xlabel("Days",fontsize=20)
axs[0,0].set_ylabel("Quantity",fontsize=20)

sns.lineplot(x="day",y="APC",data=field_data,ax=axs[0,1],color=palette[6])
axs[0,1].set_title("\nArmored Personnel Carrier\n",fontsize=25)
axs[0,1].set_xlabel("Days",fontsize=20)
axs[0,1].set_ylabel("Quantity",fontsize=20)

sns.lineplot(x="day",y="vehicles and fuel tanks",data=field_data,ax=axs[0,2],color=palette[3])
axs[0,2].set_title("\nVehicle and Fuel Tank\n",fontsize=25)
axs[0,2].set_xlabel("Days",fontsize=20)
axs[0,2].set_ylabel("Quantity",fontsize=20)

sns.lineplot(x="day",y="MRL",data=field_data,ax=axs[1,0],color=palette[0])
axs[1,0].set_title("\nMultiple Rocket Launcher\n",fontsize=25)
axs[1,0].set_xlabel("\nDays",fontsize=20)
axs[1,0].set_ylabel("Quantity",fontsize=20)

sns.lineplot(x="day",y="anti-aircraft warfare",data=field_data,ax=axs[1,1],color=palette[2])
axs[1,1].set_title("\nAnti Aircraft Warfare\n",fontsize=25)
axs[1,1].set_xlabel("\nDays",fontsize=20)
axs[1,1].set_ylabel("Quantity",fontsize=20)

sns.lineplot(x="day",y="field artillery",data=field_data,ax=axs[1,2],color=palette[1])
axs[1,2].set_title("\nField Artillery Destruction\n",fontsize=25)
axs[1,2].set_xlabel("\nDays",fontsize=20)
axs[1,2].set_ylabel("Quantity",fontsize=20)

# plt.delaxes(axs[2,0])
# plt.delaxes(axs[2,2])
plt.suptitle("Daywise Field Equipments Destruction",fontsize=30, y=1.05)
sns.despine(left=True, bottom=True)
plt.show()
```
